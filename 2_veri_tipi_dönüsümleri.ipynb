```python
pi = 3.14

yaricap = float(input("Bir değer giriniz: ")) #float olarak tanımlamak önemli çünkü 3.14 dediğimiz değer float sen bir int girersen kafa karışır o yüzden float tanımlamak önemli
daire_alan = pi * yaricap * yaricap 
daire_cevre = 2 * pi * yaricap 

print("Dairenin Alanı:",daire_alan  )
print("Dairenin çevresi:",daire_cevre)
```
