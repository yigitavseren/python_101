```python
website = "http://www.sadikturan.com"
course = "Python Kursu: Baştan sona Python Programlama Rehberiniz (40 saat)"

print(len(course))
```

```python
print(website[7:10])
```

```python
lenght = len(website)
print(website[lenght-3:lenght]) #ilginç öğren bunu
print(website[-3:])
```

```python
lenght2 = len(course)
print(course[0:15]) #14 değil 15 dikkat
print(course[lenght2-15:lenght2])
print(course[-15:]) #ikiside olur alttaki daha kolay
```

```python
print(course[::2]) #2 tane 2 nokta değişik mesela 2 yazarsan sona sadece 2şer 2şer gider
print(course[::-1]) #cevap bu babalık
```

```python
name, surname, age, job = 'Bora' , 'Yılmaz' , 32 , 'Mühendis'

print("Benim adım "+name+" "+surname+", Yaşım "+str(age) + " ve mesleğim "+job+".") #Boşluk kullanmayı unutma ve str dönüşümü yap seni mal int değil

print("Benim adım {} {} , Yaşım {} ve mesleğim {} .".format(name,surname,str(age),job)) #kesinlikle daha koaly ve pratik

print(f"Benim adım {name} {surname} , Yaşım {age} ve mesleğim {job} .") #En tatlısı bu ama bunu kullan
```
